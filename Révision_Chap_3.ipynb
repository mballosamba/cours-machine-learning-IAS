```python
import numpy as np 
import pandas as pd 
```

# 3. CREATE A DATAFRAME

```python
df = pd.DataFrame({'Date': pd.date_range(start = '2022-01-01', periods = 10, freq = 'D'), # D:fr√©quence des jours calendaires
                  'Items': 11,
                  'Mesure_1': np.random.randint(1, 10, size = 10),
                  'Mesure_2': np.random.random(10).round(2),   # round(2) .2 chiffres
                  'Mesure_3': np.random.random(10).round(2)})
```

```python
print(df)
```

```python
df.loc[[2,9], "Item"] = np.nan  # ajout de valeur manquante dans une nouvelle colonne: VM sur toutes ses lignes
df.loc[[2,7,9], "Mesure_1"] = np.nan
df.loc[[2,3], "Mesure_2"] = np.nan
df.loc[[2], "Mesure_3"] = np.nan
```

```python
print(df)
```

```python
## Supprimer toutes les lignes ayant au moins une valeur manquante(axis = 0)
d1 = df.dropna(axis=0, how='any')
d1
```

```python
## Supprimer les colonnes qui ont au moins une valeur manquante(axis = 1)
d2 = df.dropna(axis=1, how='any')
d2
```

```python
## Supprimer les lignes et colonnes qui n'ont que des valeurs manquantes (how="all")
d3 = df.dropna(axis=1, how='all')
d3
```

```python
## Delete specific rows that have missing values
d5 = df.dropna(subset = ['Mesure_1','Mesure_2']) # important, a retenir
d5
```

```python
## Replace missing values with a specfic value
values = {"Item" :1014, "Mesure_1":0} 
d6 = df.fillna(value = values) # fillna remplace les valeur manquante
d6
```

```python
## Replace missing values by mean
moyenne = df['Mesure_1'].mean()
print(moyenne)
d7 = df['Mesure_1'].fillna(df['Mesure_1'].mean())## Give a list
d7
```
